```python
#merging all three datasets
# =====================================================
# LOAD CLEAN DATASETS
# =====================================================

fire = pd.read_csv(
    "fire_mah_cleaned.csv"
)

air = pd.read_csv(
    "Mah_Daily_AirPollution_2025_cleaned_b.csv"
)

era = pd.read_csv(
    "ERA_2025_cleaned_b.csv"
)
```

```python
# =====================================================
# DATE FORMAT
# =====================================================

for df in [fire, air, era]:

    df["Date"] = pd.to_datetime(
        df["Date"]
    ).dt.date
```

```python
# =====================================================
# DISTRICT FORMAT
# =====================================================

for df in [fire, air, era]:

    df["District"] = (
        df["District"]
        .astype(str)
        .str.strip()
        .str.title()
    )
```

```python
# =====================================================
# REMOVE DUPLICATES
# =====================================================

fire = fire.drop_duplicates(
    subset=["Date", "District"]
)

air = air.drop_duplicates(
    subset=["Date", "District"]
)

era = era.drop_duplicates(
    subset=["Date", "District"]
)

print("Fire:", fire.shape)
print("Air :", air.shape)
print("ERA :", era.shape)
```

```python
# =====================================================
# MERGE AIR + FIRE
# =====================================================

merged = pd.merge(
    air,
    fire,
    on=["Date", "District"],
    how="left"
)

print(
    "\nAfter Air + Fire:",
    merged.shape
)
```

```python
# =====================================================
# MERGE ERA
# =====================================================

merged = pd.merge(
    merged,
    era,
    on=["Date", "District"],
    how="left"
)

print(
    "\nAfter ERA Merge:",
    merged.shape
)
```

```python
# =====================================================
# FINAL CHECK
# =====================================================

print("\nFINAL DATASET")

print("Rows:", merged.shape[0])
print("Columns:", merged.shape[1])

print("\nMissing Values")

print(
    merged.isnull()
    .sum()
    .sort_values(
        ascending=False
    )
)

print("\nColumns")

print(
    merged.columns.tolist()
)
```

```python
# =====================================================
# SAVE
# =====================================================

merged.to_csv(
    "Mharashtra_merged.csv",
    index=False
)

print(
    "\nSaved Maharashtra_Final_Merged_Dataset.csv"
)
```
