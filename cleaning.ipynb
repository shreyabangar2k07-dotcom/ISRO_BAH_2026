```python
import pandas as pd
import numpy as np

aqi_breakpoints = {
    "PM25": [
        (0,30,0,50),
        (31,60,51,100),
        (61,90,101,200),
        (91,120,201,300),
        (121,250,301,400),
        (251,np.inf,401,500)
    ],

    "PM10":[
        (0,50,0,50),
        (51,100,51,100),
        (101,250,101,200),
        (251,350,201,300),
        (351,430,301,400),
        (431,np.inf,401,500)
    ],

    "NO2":[
        (0,40,0,50),
        (41,80,51,100),
        (81,180,101,200),
        (181,280,201,300),
        (281,400,301,400),
        (401,np.inf,401,500)
    ],

    "SO2":[
        (0,40,0,50),
        (41,80,51,100),
        (81,380,101,200),
        (381,800,201,300),
        (801,1600,301,400),
        (1601,np.inf,401,500)
    ],

    "CO":[
        (0,1,0,50),
        (1.1,2,51,100),
        (2.1,10,101,200),
        (10.1,17,201,300),
        (17.1,34,301,400),
        (34.1,np.inf,401,500)
    ],

    "O3":[
        (0,50,0,50),
        (51,100,51,100),
        (101,168,101,200),
        (169,208,201,300),
        (209,748,301,400),
        (749,np.inf,401,500)
    ]
}
```

```python
def calculate_sub_index(concentration, pollutant):

    if pd.isna(concentration):
        return np.nan

    breakpoints = aqi_breakpoints[pollutant]

    for bp_low, bp_high, aqi_low, aqi_high in breakpoints:

        if bp_low <= concentration <= bp_high:

            sub_index = (
                ((aqi_high - aqi_low) / (bp_high - bp_low))
                * (concentration - bp_low)
                + aqi_low
            )

            return round(sub_index)

    return np.nan
```

```python
def calculate_aqi(row):

    sub_indices = []

    sub_indices.append(
        calculate_sub_index(row["PM2.5 (µg/m³)"], "PM25")
    )

    sub_indices.append(
        calculate_sub_index(row["PM10 (µg/m³)"], "PM10")
    )

    sub_indices.append(
        calculate_sub_index(row["NO2 (µg/m³)"], "NO2")
    )

    sub_indices.append(
        calculate_sub_index(row["SO2 (µg/m³)"], "SO2")
    )

    sub_indices.append(
        calculate_sub_index(row["CO (mg/m³)"], "CO")
    )

    sub_indices.append(
        calculate_sub_index(row["Ozone (µg/m³)"], "O3")
    )

    return max(sub_indices)
```

```python
df["AQI"] = df.apply(calculate_aqi, axis=1)
```

```python
df.head()
```

```python
def aqi_category(aqi):

    if aqi <= 50:
        return "Good"

    elif aqi <= 100:
        return "Satisfactory"

    elif aqi <= 200:
        return "Moderate"

    elif aqi <= 300:
        return "Poor"

    elif aqi <= 400:
        return "Very Poor"

    else:
        return "Severe"
```

```python
df["AQI_Category"] = df["AQI"].apply(aqi_category)
```
